```python
centered_price = wine_reviews.price.map(lambda p: p - (wine_reviews.price.mean()))
centered_price

#The reason this operation is slow is due to the lambda being carried out onto each element of the series.
#This is not an efficient way to perform map operations
```

```python
m = wine_reviews.price.mean()
centered_price1 = wine_reviews.price.map(lambda p: p - m)
centered_price1

#Instead, we should "vectorize" our operations when possible
# Vectorize: technique that allows operations to be applied to an entire array (vector) of data at once
```

```python
ratio = wine_reviews.points/wine_reviews.price
wine_reviews.iloc[ratio.idxmax()]
```

```python
tropical = wine_reviews.description.str.contains("tropical", case = False)
fruity = wine_reviews.description.str.contains("fruity", case = False)
print(tropical.sum())
print(fruity.sum())
```

```python
def assign_stars(row):

    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif 85 <= row.points < 95:
        return 2
    else:
        return 1
    
star_ratings = wine_reviews.apply(assign_stars, axis = 1)
star_ratings
```

```python
# groupby function doesnt actually "change" the df, the grouping is done internally.
# The important part is what type of acion you would like to do with each group (e.g. mean, mode, count, min, max)

reviews_written = wine_reviews.groupby(['taster_twitter_handle']).taster_twitter_handle.count()
reviews_written
```

```python
avg_point = wine_reviews.groupby(['title']).points.mean()
avg_point
```

```python
best_rating_by_price = wine_reviews.groupby("price").points.max()
best_rating_by_price
```

```python
# aggregate function allows the application of multiple functions on the group, very useful

variety_extremes = wine_reviews.groupby("variety").price.agg([min, max])
variety_extremes
```

```python
reviewer_mean_ratings = wine_reviews.groupby("taster_twitter_handle").points.mean()
reviewer_mean_ratings
```
